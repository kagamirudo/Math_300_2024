# Floating Point Error in Numerical Differentiation

```julia
using Plots
using LaTeXStrings
using Printf
```

```julia
default(lw=2,markersize = 6,
    xtickfont=font(12), ytickfont=font(12), 
    guidefont=font(14), titlefont=font(12))
```

Illustrate the error in approximating $f'(x)$ with the 3 point midpoint formula for $f(x) = \sin(x)$ at $x_0 =1$.

```julia
x0 = 1;
h_vals = 0.1 * 2. .^(0:-1:-30);
err_vals = [];


for h in h_vals
    fp = (sin(x0+h) - sin(x0-h))/(2*h);
    err = abs(fp - cos(x0));
    push!(err_vals, err);
    
    @printf("h = %.8e: abs. err = %g\n", h, err);
end
```

```julia
scatter(h_vals, err_vals, label="", xscale=:log10, yscale=:log10)
xlabel!(L"$h$")
ylabel!("Absolute Error")
```
